```python
dir = '../../data/hungarian.data'
with open(dir, encoding='Latin1') as file:
    lines = [line.strip() for line in file]

lines[0:10]
```

```python
import itertools
import pandas as pd

data = itertools.takewhile(
    lambda x: len(x) == 76,
    ( ' '.join(lines[i:(i+10)]).split() 
     for i in range (0, len(lines), 10))
)

df = pd.DataFrame.from_records(data)

df.head(10)
```

```python
import numpy as np

df.replace("-9", np.NaN, inplace = True)
```

```python
df.replace("-9.", np.NaN, inplace = True)

df.head(10)
```

```python
# Filling null values with 0
df.fillna(value = 0, inplace = True)

df
```

```python

df.columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal','num' ]

df.head(10)
```

```python
df['age'] = pd.to_numeric(df['age'])
df['sex'] = pd.to_numeric(df['sex'])
df['cp'] = pd.to_numeric(df['cp'])
df['trestbps'] = pd.to_numeric(df['trestbps'])
df['chol'] = pd.to_numeric(df['chol'])
df['fbs'] = pd.to_numeric(df['fbs'])
df['restecg'] = pd.to_numeric(df['restecg'])
df['thalach'] = pd.to_numeric(df['thalach'])
df['exang'] = pd.to_numeric(df['exang'])
df['oldpeak'] = pd.to_numeric(df['oldpeak'])
df['slope'] = pd.to_numeric(df['slope'])
df['ca'] = pd.to_numeric(df['ca'])
df['thal'] = pd.to_numeric(df['thal'])
df['num'] = pd.to_numeric(df['num'])

df.info()
```
